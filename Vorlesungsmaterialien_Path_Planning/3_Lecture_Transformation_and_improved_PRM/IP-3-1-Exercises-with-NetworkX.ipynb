# Exercises with NetworkX

1. Graph with x nodes at random position
2. Graph with x nodes at random position and all nodes are connected with edges
3. Graph with x nodes at random position, nodes with given distance are connected


Following the results from the lecture discussion


```python
import math
import networkx as nx
import random
import matplotlib.pylab as plt
```

```python
G_new = nx.Graph()
num_nodes = 50
min_dist = 40

def dist(a,b):
    return math.sqrt((b["pos"][0]-a["pos"][0])**2 + (b["pos"][1]-a["pos"][1])**2)

for i in range(num_nodes):
    G_new.add_node(i, pos = [random.uniform(0,100),random.uniform(0,100)])

pos = nx.get_node_attributes(G_new,"pos")

for na in G_new.nodes:
    for nb in G_new.nodes:
        if na!=nb:
            if dist(G_new.nodes[na],G_new.nodes[nb]) <= min_dist:
                G_new.add_edge(na,nb)

fig, axes = plt.subplots()

nx.draw(G_new, pos=pos, with_labels=True)          
```
